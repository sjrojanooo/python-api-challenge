```python
#loading the csv that was exported in part 1 to a dataframe 
weather_file = 'City_Data.csv'

#reading the data in the csv 
weather_df = pd.read_csv(weather_file,encoding='utf-8')

#Creating the DataFrame 
weather_df = pd.DataFrame(weather_df)

#Displaying the DataFrame
weather_df
```

Humidity Heatmap

Configure gmaps.
Use the Lat and Lng as locations and Humidity as the weight.
Add Heatmap layer to map.

```python
#configuring gmaps with an API key 
gmaps.configure(api_key=g_key)

#Storing the lat and long into locations 
locations = weather_df[["Latitude","Longtitude"]].astype(float)

humidity = weather_df["Humidity"]
```

```python
#Creating a humidity head map 
figure_layout = {
    'width': '900px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}

fig = gmaps.figure(layout=figure_layout,zoom_level=2,center=(15,25))

heat_map = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=1.5)

fig.add_layer(heat_map)

fig
```

Create new DataFrame fitting weather criteria

Narrow down the cities to fit weather conditions.
Drop any rows will null values.

```python
#Filtering out the data where the max temp is above 70 and below 80 
perf_weather = weather_df[(70.0 < weather_df['Max Temp']) & (weather_df['Max Temp'] < 80.0)]

clean_df = perf_weather

clean_df
```

```python
#Filtering out the data frame data to show where the wind speed is above 0 mph and below 10 mph
perf_wind = clean_df[(0 < clean_df['Wind Speed']) & (clean_df['Wind Speed'] < 10.0) & (clean_df['Cloudiness'] == 0)]

clean_df = perf_wind

clean_df

#We are now left with locations where all three conditions are met. Showing the most 
#ideal places to have a vacation 
hotel_df = pd.DataFrame(clean_df)

hotel_df
```

```python
#Creating another heat map for the fun of it. 
perfect_hum = hotel_df['Humidity']

perfect_loc = hotel_df[['Latitude','Longtitude']].astype(float)


#Creating a humidity head map 
figure_layout = {
    'width': '900px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}

fig = gmaps.figure(layout=figure_layout,zoom_level=2,center=(15,25))

heat_map = gmaps.heatmap_layer(perfect_loc, weights=perfect_hum, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=1.5)

fig.add_layer(heat_map)

fig
```

```python
#adding hotel name column to dataframe 
hotel_df["Hotel Name"] = "" 

hotel_df = hotel_df.rename(columns={"Latitude":"Lat","Longtitude":"Lng"})

hotel_df = hotel_df[['City','Country','Lat','Lng','Hotel Name']]

hotel_df
```

```python
#creating the base url to send the api requests to 
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

#Configuring my api settings 
params = {"type" : "hotel",
          "keyword" : "hotel",
          "radius" : 5000,
          "key" : g_key}
```

```python
#initating my for loop 
for index, row in hotel_df.iterrows():
    
# get city name, lat, lnt from df
    lat = row["Lat"]
    lng = row["Lng"]
    city_name = row["City"]
    
    # add keyword to params dict
    params["location"] = f"{lat},{lng}"

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {city_name}.")
    response = requests.get(base_url, params=params).json()
    
    # getting results from the json response 
    results = response['results']
    
    # save the hotel name to dataframe
    try:
        print(f"Closest hotel in {city_name} is {results[0]['name']}.")
        hotel_df.loc[index, "Hotel Name"] = results[0]['name']

    #allows program to keep searching even if there isn't a match 
    #Instead of quitting the returned value will be this print out 
    except (KeyError, IndexError):
        print("No Hotel within a 5000 meter radius")


# Print end of search once searching is completed
print("*****API REQUESTS ARE FINISHED******")
```

```python
#replacing blank values with NaN
hotel_df['Hotel Name'].replace('', np.nan, inplace=True)

#Removing locations without a nearby hotel from the data set 
hotel_df.dropna(subset=['Hotel Name'],inplace=True)

#Displaying the new data set 
hotel_df
```

```python
 # NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
 # Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content = hotel_info)

# Add the layer to the map
fig.add_layer(markers)

# Display Map
fig
```
